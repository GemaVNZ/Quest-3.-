```python
url = "https://api.openchargemap.io/v3/poi/"

params_base = {
    "output": "json",
    "maxresults": 10000, 
    "compact": "true", 
    "verbose": "false", 
    "key": "d1138da0-c7d4-4f81-a2c6-44de7bbd80a0"

}

country_codes = ["ES", "FR"]

data_country = []

for country in country_codes:
    params = params_base.copy()
    params["countrycode"] = country
    
    response = requests.get(url, params=params)

    if response.status_code == 200:
       data = response.json()  
       data_country.extend(data)
    else:
        print(f"Error en la solicitud para {country}: {response.status_code}")
    
data_chargesmap= pd.DataFrame(data_country)
```

```python
data_chargesmap.drop(['IsRecentlyVerified','DateLastVerified','UUID','DataProviderID','OperatorID','UsageTypeID','NumberOfPoints','StatusTypeID','DateLastStatusUpdate','DataQualityLevel','DateCreated','SubmissionStatusTypeID','GeneralComments','DataProvidersReference','OperatorsReference'], axis=1, inplace=True)
```

```python
data_chargesmap['Address'] = data_chargesmap['AddressInfo'].apply(
    lambda x: f"{x.get('AddressLine1', '')}, {x.get('AddressLine2', '')}".strip(', ') if x else None)
data_chargesmap['Title'] = data_chargesmap['AddressInfo'].apply(lambda x: x.get('Title') if x else None)
data_chargesmap['Region'] = data_chargesmap['AddressInfo'].apply(lambda x: x.get('StateOrProvince') if x else None)
data_chargesmap['City'] = data_chargesmap['AddressInfo'].apply(lambda x: x.get('Town') if x else None)
data_chargesmap['PostalCode'] = data_chargesmap['AddressInfo'].apply(lambda x: x.get('Postcode') if x else None)
data_chargesmap['CountryID'] = data_chargesmap['AddressInfo'].apply(lambda x: x.get('CountryID') if x else None)
```

```python
data_chargesmap = data_chargesmap.reindex(columns=['ID','UsageCost','Address','Title', 'City','Region','PostalCode','CountryID', 'Connections','AddressInfo']) 
```
